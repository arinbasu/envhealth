What is Air Pollution?
----------------------
- Mix of natural and man-made substances in our breathing air
- Two categories: outdoor and indoor

## Outdoor Air Pollution
- Exposure outside of built environment
- Burn fossil fuels - fine particulate matter
- Noxious gases (SO2, Nitrogen Oxides, Carbon Monoxide, Chemical Vapour)
- Ground Level Ozone
- Tobacco Smoke

## Indoor Air Pollution
- Particulate matter, carbon oxides, polllutants in indoor air or dust
- Gases (Radon and CO)
- Household products
- Indoor allergens
- Secondhand smoke, molds, pollens

## Main effects of air Pollution
- Asthma
- Respiratory function changes
- Pregnancy (Preterm birth)
- Cancer

## Acute health effects of Air Pollution
- Meuse Valley (1930) - 65 deaths
- Donora, PA (1948) - 20 deaths
- London, UK (1952), 4000 deaths

## Why are these important?
- Wake up call to the world in their time
- Air pollution is more than "nuisance"

## London Smog: 
- Road, rail, air transport got to a standstill
- Performance at Sadler wells theatre stalled as the conditions became unbearable
- At a cattle show, the cattle were asphyxiated
- People could not see their own feet
- Fog so dense they called it "Pea souper"
- Caused due to factory emissions from noxious gases

## Sources of Air Pollution: harms to children
- Power plants, vehicles, factories - spew harmful gases
- Small particles
- Penetrate deeply into little children's lungs
- Oxides of Nitrogen + Sunlight -> Ozone
- Ozone can trigger Asthma attacks

## Cross Border Pollution
- Wind carries pollutants across boundaries
- Indonesian Forest Fires
- Haze of Smoke
- Schools were forced to close in neighbouring countries
- Problems are faced worse in meagacities

## Burden of Air Pollution
- Spectrum of disease and conditions
- Eye irritation to death

## Quantifying the health effects is a challenge
- Limited information on health effects
- Limited information on air pollution in many parts of the world
- Most air pollution comes from combustion processes

## Particulate Matters
- PM is consistently associated with serious health effects
- Deaths from heart and lung diseases in adults
- Deaths from lung cancer
- Deaths from acute respiratory illnesses in children
- Outdoor PM pollution responsible for 3% of adult deaths due to cardiac and pulmonary diseases
- 5% of lung cancer related deaths
- 1% deaths in children due to acute respiratory conditions
- The burden of disease occurs mainly in developing countries

What are the main sources of Air Pollutants?
----------------------------

- Differs between rural and urban settings
- Combustion of fossil fuels
- Car Exhaust
- Volcanic Eruptions (NZ: White Island)

## Factors that affect air pollution
- Climate 
- Topography
- Can be troublesome if city is in a valley surrounded by Mountains (think of Waimakariri Valley, or Canterbury)
- Roads can generate air pollution when people drive cars on them

## Inversion Layer
- Normally, we have warm air below closer to the ground
- Cold air resides above the warm layer
- The warm layer gradually rises and the cold air subsides
- In this situation, the warm air is trapped above the cold air
- This traps dangerous concentrations in the cooler air below
- Cause of dense smog in urban areas

## Why are children particularly vulnerable to air pollution? 
- Children's lung function are immature
- They are shorter in height
- Therefore exposed to air pollutants in higher concentrations
- The younger a child, the worse it gets as till about 8 years of age, children's lungs do not mature enough
- Higher rate of oxygen consumption per unit body weight
- They have narrower airways

## Major outdoor pollutants ("Big five")
- Particulate Matter
- Ozone
- Oxides of Nitrogen
- Carbon Monoxide
- Sulphur Oxide

## Particulate Matter
- Complex mixture of particulate
- Solid, liquid, or both
- Power plants
- Industry
- Volcanoes, dust storms
- Carbon cores with chemicals adsorbed on it

## PM composition
- Varies by region, weather conditions, others
- AKA Total suspended particulates
- Black smoke
- RSPM (Respirable particulate matter)
- PM10 or PM2.5 (ultrafine)
- PM10 or PM2.5 settle in upper airways
- PM less than 2.5 travel deeper in the respiratory tree

## Health Effects of Particulate Matters
- Pneumonia
- Asthma
- Increased deaths from all causes
- Decreased lung function

## In general we know that
- For every 10 ug/L increase in PM10,
- Total death rates increase by 1%
- Heart disease related death rates increase by 1.4%
- Respiratory disease related death rates by 3.4%
- Asthma related hospitalisations 1.9%
- Asthma worsenings 3%
- Asthma related emergency room visits 3.4%

## Sources of Ozone
- Volatile Organics + Oxygen = Ozone
- Oxides of Nitrogen + Oxygen = Ozone
- Peaks late afternoon
- Maximum in hot stagnant air
- Good Ozone is in upper atmosphere
- That Ozone shields from UV radiation

## Bad Ozone
- Occurs in Lower Atmosphere
- Major component of urban smog
- Respiratory irritant
- Increases sensitivity to other allergens
- Tends to peak between 3-5 PM 
- Requires sunlight and heat
- Schedule your vigorous activity (runs) in early morning or after dark
- Know your air quality index and adjust your evening runs
- Damages lungs

## Air Quality Index
- Measured by five components: PM, Ozone, oxides of N, sulphur, CO
- From 0-500
- The greater the value, the worse the situation
- The greater the value, the more the health concern
- A value of 50 or lower is good
- A value of 100 is generally National Air Quality Standard
- Anything above 100 bad

## Oxides of Nitrogen
- Power plants and industry
- Motor Vehicles
- Natural sources (volcanoes, lightning, bacteria)
- Forms Ozone
- Decreases lung function
- Increases airway reactivity
- Increased susceptibility to infections

## Carbon Monoxide
- Motor Vehicle Exhausts Most Important Source
- Fires
- Industry Exhaust
- Wood fire
- Combines with haemoglobin in blood
- Dangerous and can cause death
- Severely reduces Oxygen carrying capacity

## Sulphur Oxides
- Combustion of sulphur containing coal or oil
- Industrial processes
- volcanoes
- Can combine with water to form acids
- Acid Rain
- Airway diseases
- decreased lung function

## Air Pollutants that can cause cancer
- Arsenic and lung cancer
- Benzene and leukaemia
- Chromium and lung cancer
- Vinyl Chloride and liver cancer
- Radon and lung cancer
- PAH and lung cancer
- Majority of them can cause lung cancer
- Fluorosis can be caused by burning high fluorine containing coal (problem in China)

## Biological Pollutants
- Moulds
- Biological materials such as soya bean dusts
- Pollens
- Air conditioners and diseases
- Need for High Efficiency Particulate Arrestance (HEPA) filters

## What can we do about it?
- Replace coal as fuel
- Change transportation (use public transportation)
- Adhere to air quality guidelines
